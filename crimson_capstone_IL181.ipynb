```python
#cell input hide code
#taken from: https://newbedev.com/how-to-hide-one-specific-cell-input-or-output-in-ipython-notebook

from IPython.display import HTML
import random

def hide_toggle(for_next=False):
    this_cell = """$('div.cell.code_cell.rendered.selected')"""
    next_cell = this_cell + '.next()'

    toggle_text = 'Toggle show/hide'  # text shown on toggle link
    target_cell = this_cell  # target cell to control with toggle
    js_hide_current = ''  # bit of JS to permanently hide code in current cell (only when toggling next cell)

    if for_next:
        target_cell = next_cell
        toggle_text += ' next cell'
        js_hide_current = this_cell + '.find("div.input").hide();'

    js_f_name = 'code_toggle_{}'.format(str(random.randint(1,2**64)))

    html = """
        <script>
            function {f_name}() {{
                {cell_selector}.find('div.input').toggle();
            }}

            {js_hide_current}
        </script>

        <a href="javascript:{f_name}()">{toggle_text}</a>
    """.format(
        f_name=js_f_name,
        cell_selector=target_cell,
        js_hide_current=js_hide_current, 
        toggle_text=toggle_text
    )

    return HTML(html)

hide_toggle()
```

# 4. Running the Models

```python
#split data
X = df_final_dummies.drop("rebuy_182days", axis=1)
Y = df_final_dummies["rebuy_182days"].astype('int64')
```

### feature selection through mutual information, before fitting the models

```python
from sklearn.feature_selection import mutual_info_classif as MIC

#compute the mutual info scores
mi_score = MIC(X,Y, discrete_features=False)
```

```python
#constructing a dataframe that maps the mi scores with the X_columns
d = {'X_columns':list(X.columns), 'mi_scores': list(mi_score)}
mi_scores_map = pd.DataFrame(d)

mi_scores_map.sort_values('mi_scores',ascending=False).head(15)
```

```python
plt.hist(mi_scores_map['mi_scores'], bins=50)
plt.xlabel('Threshold')
plt.ylabel('Number of Features')
```

```python

#minimum threshold for the mutual information scores to be selected
thresholds = [0, 0.001, 0.003, 0.005, 0.007, 0.01, 0.015, 0.02, 0.05]

for i in thresholds:
    mi_scores_map_selected = mi_scores_map[mi_scores_map['mi_scores']>i]

    X_2 = X[mi_scores_map_selected['X_columns']]
    X_train_2, X_test_2, Y_train_2, Y_test_2 = train_test_split(X_2, Y, test_size=0.20, 
                                                                random_state=42)

    random_forest = RandomForestClassifier(n_estimators=100)
    random_forest.fit(X_train_2, Y_train_2)
    Y_pred_train_2 = random_forest.predict(X_test_2)
    Y_pred_test_2 = random_forest.predict(X_test_2)
    acc_random_forest = round(random_forest.score(X_test_2, Y_test_2) * 100, 2)


    print("Threshold: ", i, " Accuracy Score: " acc_random_forest)
```
